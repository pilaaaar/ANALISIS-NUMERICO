### Load Data and Preprocessing

se eliminan las columnas Participante y EMG

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/DM.csv')

# Drop the 'EMG' and 'Participante' columns
df_processed = df.drop(columns=['EMG', 'Participante'])

print("Original DataFrame head:")
display(df.head())

print("\nProcessed DataFrame head (after dropping 'EMG' and 'Participante'):")
display(df_processed.head())

print("\nInformation about the processed DataFrame:")
df_processed.info()
```

### Prepare Data for Classification



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Define features (X) and target (y)
X = df_processed.drop(columns=['Grupo'])
y = df_processed['Grupo']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Standardize the features (important for neural networks and SVMs)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"X_train_scaled shape: {X_train_scaled.shape}")
print(f"X_test_scaled shape: {X_test_scaled.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

### Multilayer Perceptron (MLP) Classifier

Entrenamos un Multilayer Perceptron (MLP) classifier en la  training data usando `accuracy_score`, `classification_report` `confusion_matrix`para ver su efectividad

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize and train the MLP Classifier
mlp_classifier = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42, activation='relu', solver='adam')
mlp_classifier.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred_mlp = mlp_classifier.predict(X_test_scaled)

# Evaluate the MLP model
print("\nMLP Classifier Evaluation:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_mlp):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_mlp))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_mlp))
```

### Support Vector Machine (SVM) Classifier



```python
from sklearn.svm import SVC

# Initialize and train the SVM Classifier
svm_classifier = SVC(kernel='rbf', random_state=42) # Using Radial Basis Function (RBF) kernel
svm_classifier.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred_svm = svm_classifier.predict(X_test_scaled)

# Evaluate the SVM model
print("\nSVM Classifier Evaluation:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_svm):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_svm))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_svm))
```

### Exporting Classification Reports to CSV


```python
import pandas as pd
from sklearn.metrics import classification_report

def get_classification_report_df(y_true, y_pred, model_name):
    report_dict = classification_report(y_true, y_pred, output_dict=True)
    df_report = pd.DataFrame(report_dict).transpose()
    df_report['model'] = model_name
    df_report = df_report.reset_index().rename(columns={'index': 'metric'})
    return df_report

# Get report for MLP
mlp_report_df = get_classification_report_df(y_test, y_pred_mlp, 'MLP Classifier')

# Get report for SVM
svm_report_df = get_classification_report_df(y_test, y_pred_svm, 'SVM Classifier')

# Combine the reports
combined_report_df = pd.concat([mlp_report_df, svm_report_df], ignore_index=True)

# Save to CSV
csv_filename = 'classification_report_summary.csv'
combined_report_df.to_csv(csv_filename, index=False)

print(f"Classification reports saved to {csv_filename}")
display(combined_report_df)
```

### Hyperparameter Tuning for MLP Classifier with Cross-Validation

I will use `GridSearchCV` to find the best hyperparameters for the MLP Classifier. This process involves defining a grid of parameters to search over and using 5-fold cross-validation to evaluate each combination, ensuring a more robust parameter selection.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier # Import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix # Import for evaluation

# Define the parameter grid for MLP
mlp_param_grid = {
    'hidden_layer_sizes': [(100,), (50, 50), (100, 50, 25)],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant', 'adaptive'],
}

# Initialize GridSearchCV with 5-fold cross-validation
mlp_grid_search = GridSearchCV(
    MLPClassifier(max_iter=500, random_state=42),
    mlp_param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

# Fit the grid search to the scaled training data
print("Starting GridSearchCV for MLP (with 5-fold cross-validation)...")
mlp_grid_search.fit(X_train_scaled, y_train)

# Print the best parameters and best score found through cross-validation
print("\nBest parameters for MLP:", mlp_grid_search.best_params_)
print("Best cross-validation accuracy for MLP:", mlp_grid_search.best_score_)

# Evaluate the best MLP model on the independent test set
best_mlp = mlp_grid_search.best_estimator_
y_pred_best_mlp = best_mlp.predict(X_test_scaled)

print("\nMLP Classifier Evaluation (on test set after tuning):")
print(f"Accuracy: {accuracy_score(y_test, y_pred_best_mlp):.4f}")
print("Classification Report:")
print(classification_report(y_test, y_pred_best_mlp))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_best_mlp))
```

### Hyperparameter Tuning for SVM Classifier with Cross-Validation

Next, I will perform hyperparameter tuning for the SVM Classifier using `GridSearchCV`, again with 5-fold cross-validation, to find the optimal `C` and `gamma` values for the 'rbf' kernel.

```python
from sklearn.svm import SVC # Import SVC
from sklearn.model_selection import GridSearchCV # Ensure GridSearchCV is imported for consistency
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix # Import for evaluation

# Define the parameter grid for SVM
svm_param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf']
}

# Initialize GridSearchCV with 5-fold cross-validation
svm_grid_search = GridSearchCV(
    SVC(random_state=42),
    svm_param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

# Fit the grid search to the scaled training data
print("Starting GridSearchCV for SVM (with 5-fold cross-validation)...")
svm_grid_search.fit(X_train_scaled, y_train)

# Print the best parameters and best score found through cross-validation
print("\nBest parameters for SVM:", svm_grid_search.best_params_)
print("Best cross-validation accuracy for SVM:", svm_grid_search.best_score_)

# Evaluate the best SVM model on the independent test set
best_svm = svm_grid_search.best_estimator_
y_pred_best_svm = best_svm.predict(X_test_scaled)

print("\nSVM Classifier Evaluation (on test set after tuning):")
print(f"Accuracy: {accuracy_score(y_test, y_pred_best_svm):.4f}")
print("Classification Report:")
print(classification_report(y_test, y_pred_best_svm))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_best_svm))
```

### Explicit Cross-Validation on Best Models

To explicitly show the cross-validation results for the best MLP and SVM models (after tuning), I'll use `cross_val_score` on the *entire* dataset for a final, generalizable performance estimate. This demonstrates how well the models would perform on unseen data across multiple splits.

```python
from sklearn.model_selection import cross_val_score

# Get the best estimators from GridSearchCV
best_mlp_model = mlp_grid_search.best_estimator_
best_svm_model = svm_grid_search.best_estimator_

# Combine scaled training and test data for full dataset cross-validation
X_combined_scaled = scaler.fit_transform(X) # Re-scale the full dataset
y_combined = y

# Perform 5-fold cross-validation for the best MLP model
mlp_cv_scores = cross_val_score(best_mlp_model, X_combined_scaled, y_combined, cv=5, scoring='accuracy', n_jobs=-1)
print("\nCross-validation scores for Best MLP Model (5-fold):\n", mlp_cv_scores)
print(f"Mean CV accuracy for Best MLP Model: {mlp_cv_scores.mean():.4f}")
print(f"Standard deviation of CV accuracy for Best MLP Model: {mlp_cv_scores.std():.4f}")

# Perform 5-fold cross-validation for the best SVM model
svm_cv_scores = cross_val_score(best_svm_model, X_combined_scaled, y_combined, cv=5, scoring='accuracy', n_jobs=-1)
print("\nCross-validation scores for Best SVM Model (5-fold):\n", svm_cv_scores)
print(f"Mean CV accuracy for Best SVM Model: {svm_cv_scores.mean():.4f}")
print(f"Standard deviation of CV accuracy for Best SVM Model: {svm_cv_scores.std():.4f}")
```
